```python
import pandas as pd
import numpy as np
```

```python
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')
```

```python
print(file1.shape)
print(file2.shape)
print(file3.shape)
```

```python
file1.head()
```

```python
file2.head()
```

```python
file3.head()
```

```python
file1= file1.rename(columns={'ST':'State', 'GENDER':'Gender',})
file2= file2.rename(columns={'ST':'State', 'GENDER':'Gender',})

#file2.head()
```

```python
file1=file1[['Customer', 'State','Gender','Education', 'Customer Lifetime Value', 'Income', 'Monthly Premium Auto', 'Number of Open Complaints', 'Total Claim Amount', 'Policy Type', 'Vehicle Class']]
file2=file2[['Customer', 'State','Gender','Education', 'Customer Lifetime Value', 'Income', 'Monthly Premium Auto', 'Number of Open Complaints', 'Total Claim Amount', 'Policy Type', 'Vehicle Class']]
```

```python
file1.head()
```

```python
file1['Customer Lifetime Value']=file1['Customer Lifetime Value'].str.replace('%','') 
file2['Customer Lifetime Value']=file2['Customer Lifetime Value'].str.replace('%','')

file1['Customer Lifetime Value']=pd.to_numeric(file1['Customer Lifetime Value'], errors='coerce')
file2['Customer Lifetime Value']=pd.to_numeric(file2['Customer Lifetime Value'], errors='coerce')
file3['Customer Lifetime Value']=pd.to_numeric(file3['Customer Lifetime Value'], errors='coerce')

file1['Customer Lifetime Value']=file1['Customer Lifetime Value']/100
file2['Customer Lifetime Value']=file2['Customer Lifetime Value']/100

all_data = pd.concat([file1, file2, file3], axis=0)
display(all_data)
```

```python
print(all_data.dtypes)
```

### Which columns are numerical?

Customer Life value, Income, Monthly Premium Auto, Number of Open Complaints, Total Claim Amount

```python
all_data._get_numeric_data()
```

### Which columns are categorical?

Customer, State, Gender, Education, Policy Type, Vehicle Class

```python
all_data.select_dtypes('object')
```

# Data Cleaning

```python
all_data = all_data.drop(['Education', 'Number of Open Complaints'], axis=1)
display(all_data.head())
```
